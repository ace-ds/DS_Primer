#### Data Granularity 

Here, it seems that each row represent a particular event that occured in a country

````python
country_events = country_events.drop(['id'], axis=1)
country_events = country_events[country_events.year.isin(['2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015'])]
````

````python

for index, row in country_events.iterrows():
    row['country'] = (row['country']).strip()
    row['year'] = (row['country']).strip()
        
country_events.loc[country_events['country'] == 'Yemen (North Yemen)', 'country'] = 'Yemen'
country_events.loc[country_events['country'] == 'DR Congo (Zaire)', 'country'] = 'DR Congo'
country_events.loc[country_events['country'] == 'Madagascar (Malagasy)', 'country'] = 'Madagascar'
country_events.loc[country_events['country'] == 'Zimbabwe (Rhodesia)', 'country'] = 'Zimbabwe'
country_events.loc[country_events['country'] == 'Serbia (Yugoslavia)', 'country'] = 'Serbia'
country_events.loc[country_events['country'] == 'Cambodia (Kampuchea)', 'country'] = 'Cambodia'
country_events.loc[country_events['country'] == 'Russia (Soviet Union)', 'country'] = 'Russia'
country_events.loc[country_events['country'] == 'Myanmar (Burma)', 'country'] = 'Myanmar'
country_events['year'] = country_events['year'].apply(str)
country_events['country'] = country_events['country'].apply(str)

country_events.to_csv('news_article_conflicts.csv')

````

````python
# Forming Conflict Aggregate File
conflict_aggregate = country_events.groupby(['country','year'], as_index=False).agg({'best': 'sum', 'deaths_civilians': 'sum'})
conflict_aggregate=conflict_aggregate.rename(columns = {'best':'total_deaths'})
conflict_aggregate=conflict_aggregate.rename(columns = {'deaths_civilians':'civilian_deaths'})
conflict_aggregate.to_csv('conflict_aggregate.csv')
````

````python
#Creating the Human Rights Master 
import numpy as np
human_rights = pd.read_csv('../Dataset/CIRI_Data.csv')
human_rights = human_rights.replace(-77,np.NaN)
human_rights = human_rights.replace(-66,np.NaN)
human_rights = human_rights.replace(-999,np.NaN)
````

````python

human_rights = human_rights.dropna(thresh=0.7*len(human_rights), axis=1)
````

````python
human_rights.columns
human_rights.isnull().sum()
````

````python
human_rights['PHYSINT'].fillna((human_rights['PHYSINT'].mode()[0]), inplace=True)
human_rights['DISAP'].fillna((human_rights['DISAP'].mode()[0]), inplace=True)
human_rights['KILL'].fillna((human_rights['KILL'].mode()[0]), inplace=True)
human_rights['POLPRIS'].fillna((human_rights['POLPRIS'].mode()[0]), inplace=True)
human_rights['TORT'].fillna((human_rights['TORT'].mode()[0]), inplace=True)
human_rights['NEW_EMPINX'].fillna((human_rights['NEW_EMPINX'].mode()[0]), inplace=True)
human_rights['ASSN'].fillna((human_rights['ASSN'].mode()[0]), inplace=True)
human_rights['FORMOV'].fillna((human_rights['FORMOV'].mode()[0]), inplace=True)
human_rights['DOMMOV'].fillna((human_rights['DOMMOV'].mode()[0]), inplace=True)
human_rights['SPEECH'].fillna((human_rights['SPEECH'].mode()[0]), inplace=True)
human_rights['ELECSD'].fillna((human_rights['ELECSD'].mode()[0]), inplace=True)
human_rights['NEW_RELFRE'].fillna((human_rights['NEW_RELFRE'].mode()[0]), inplace=True)
human_rights['WORKER'].fillna((human_rights['WORKER'].mode()[0]), inplace=True)
human_rights['WECON'].fillna((human_rights['WECON'].mode()[0]), inplace=True)
human_rights['WOPOL'].fillna((human_rights['WOPOL'].mode()[0]), inplace=True)
human_rights['INJUD'].fillna((human_rights['INJUD'].mode()[0]), inplace=True)
human_rights = human_rights.drop(['COW', 'POLITY','UNCTRY'], axis=1)
````
